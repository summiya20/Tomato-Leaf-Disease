```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
tf.__version__
```

```python
train_datgen = ImageDataGenerator(
        rescale = 1./255,
        shear_range = 0.2,
        zoom_range = 0.2,
        horizontal_flip = True)

train_set = train_datgen.flow_from_directory(
        '/kaggle/input/tomatoleaf/tomato/train',
        target_size =(64, 64),
        batch_size = 32,
        class_mode ='categorical')
```

```python
test_datagen = ImageDataGenerator(rescale = 1./255)
test_set = test_datagen.flow_from_directory(
         '/kaggle/input/tomatoleaf/tomato/val',
        target_size =(64, 64),
        batch_size = 32,
        class_mode ='categorical'),
```

```python
print(f'Class names:{train_set.class_indices}')
```

```python
cnn = tf.keras.models.Sequential()
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=[64, 64, 3]))
```

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tf.keras.layers.Flatten())
```

```python
cnn.add(tf.keras.layers.Dense(units=128, activation='relu'))
```

```python
cnn.add(tf.keras.layers.Dense(units=10, activation='softmax'))
```

```python
cnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
cnn.fit(x = train_set, validation_data = test_set, epochs = 20)
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image

# Define your class names
class_names = ['Tomato___Bacterial_spot', 'Tomato___Early_blight', 'Tomato___Late_blight', 'Tomato___Leaf_Mold', 'Tomato___Septoria_leaf_spot', 'Tomato___Spider_mites Two-spotted_spider_mite', 'Tomato___Target_Spot', 'Tomato___Tomato_Yellow_Leaf_Curl_Virus', 'Tomato___Tomato_mosaic_virus', 'Tomato___healthy']

# Assuming 'cnn' is already defined and loaded with the trained model

test_image = image.load_img('/kaggle/input/single-prediction/potato-blight.jpg', target_size=(64, 64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis=0)
result = cnn.predict(test_image)
predicted_class_index = np.argmax(result)
predicted_class_name = class_names[predicted_class_index]
print("Predicted Class:", predicted_class_name)
```
